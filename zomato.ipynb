```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_country = pd.read_csv('Country-Code.csv')
df_country.head()
```

```python
df.columns
```

```python
final_df = pd.merge(df,df_country,on='Country Code',how='left')
```

```python
final_df.head()
```

```python
final_df.columns
```

```python
final_df.Country.value_counts()
```

```python
country_names = final_df.Country.value_counts().index
```

```python
country_val = final_df.Country.value_counts().values
```

```python
plt.pie(country_val[:3],labels = country_names[:3],autopct = "%1.2f%%") #[:3] shows top 3 country name and their values
```

OBSERVATION:  maximum  records or transaction in zomato are done by INDIA then UNITED STATES and then UNITED KINGDOM

```python
final_df.columns
```

```python
ratings = final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'rating count'})
```

```python
ratings
```

## observation
1. when rating is between 4.5 to 4.9 ----> excellent
2. when ratings are between 4.0 to 4.4 ----> very good
3. when ratings are between 3.5 to 3.9 ----> good
4. when ratings are between 2.5 to 3.4 ----> average
5. when ratings are between 2.0 to 2.4 ----> poor

```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,6)
sns.barplot(x='Aggregate rating',y = 'rating count',data = ratings)
```

```python
sns.barplot(x='Aggregate rating',y = 'rating count',hue = 'Rating color',data = ratings,palette = ['white','red','orange','yellow','green','green'])
```

```python
sns.countplot(x = "Rating color",data = ratings,palette = ['blue','red','orange','yellow','green','green'])
```
